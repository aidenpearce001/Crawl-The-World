```python
import os

def getImgUrl(user_entry):
    imgUrl = []
    # 1 cuz zero idx means the owner of this tweet
    for idx in range(1, len(user_entry)):
        print("Get images from user {}".format(idx))
        try:
            legacy = user_entry[idx]["content"]["itemContent"]["tweet"]["legacy"]
#             print(legacy)
            print("Title: ", legacy["full_text"])
        except KeyError as e:
            print(e)
            continue
        if("entities" in legacy):
            imgUrl = getUrlFromEntity(legacy, imgUrl)
        if("retweeted_status" in legacy):
            imgUrl = getUrlFromRetweeted(legacy, imgUrl)
        
        
    return imgUrl

def getUrlFromEntity(legacy, imgUrl):
    entities = legacy["entities"]
    try:
        medias = entities["media"]
    except KeyError as e:
        return imgUrl
    
    for media in medias:
            if("media_url_https" in media):
                imgUrl.append(media["media_url_https"])
            else:
                continue
    return imgUrl

def getUrlFromRetweeted(legacy, imgUrl):
    entities = legacy["retweeted_status"]["legacy"]["entities"]
#     print(entities)
    try:
        medias = entities["media"]
    except KeyError as e:
        return imgUrl    
    for media in medias:
        if("media_url_https" in media):
            imgUrl.append(media["media_url_https"])
        else:
            continue
    return imgUrl

def downloads(url, folder):
    img_name = url.split("/")[-1]
    print("Download image ", img_name)
    img_data = requests.get(url).content
    saveFolder = os.path.join(folder, img_name)
    with open(saveFolder, 'wb') as handler:
        handler.write(img_data)
        
def downloadImgFromUrl(imgUrl):
    folder = "images"
    print("Start downloading ...")
    for url in imgUrl:
        downloads(url, folder)
    numCrawledImages = len(os.listdir(folder))
    
#     assert numCrawledImages == len(imgUrl), "Some urls wrong"
    print(f"Finish download {numCrawledImages} url")
```

```python
imgUrl = getImgUrl(user_entry)
downloadImgFromUrl(imgUrl)
```
